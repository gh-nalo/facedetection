# Convert List Of Bboxes To mAP-needed Format

```python
import os
```

### Convert Yolo Format to mAP needed format

```python
# <class_name> <left> <top> <right> <bottom>
```

```python
base_path = "./dataset_augmented/train"

label_path = os.path.join(base_path, "labels")
new_label_path = os.path.join(base_path, "labels_new")
image_path = os.path.join(base_path, "images")

for idx, textfile in enumerate(os.listdir(label_path)):
    print(textfile)

    file_path = os.path.join(label_path, textfile)

    strings_to_write = []

    # Read text file
    with open(file_path, 'r') as f:
        lines = f.readlines()

    for line_idx, unstripped_line in enumerate(lines):
        line = unstripped_line.strip()
        split = [float(x) for x in line.split(" ")]
        # print(line)

        # In case image width is not scaled to 640, 640 already
        # image_name = textfile[0:-4] + ".jpg"
        # jpg_image_path = os.path.join(image_path, image_name)
        # print(jpg_image_path)
        # image = cv.imread(jpg_image_path)

        centerX = split[1]
        centerY = split[2]
        width = split[3]
        height = split[4]

        # Top left corner
        x1 = centerX - (width / 2.0)
        y1 = centerY - (height / 2.0)
        x2 = (x1 + width)
        y2 = (y1 + height)

        # Re-calculate coordinates
        className = "Face"  # split[0]
        x1 = int(x1 * 640)
        y1 = int(y1 * 640)
        x2 = int(x2 * 640)
        y2 = int(y2 * 640)

        lineToWrite = ' '.join([str(x) for x in [className, x1, y1, x2, y2]])
        # print(lineToWrite)
        strings_to_write.append(lineToWrite)

    print(strings_to_write)

    # Write new text file with same name in labels_new directory
    new_file_path = os.path.join(new_label_path, textfile)
    with open(new_file_path, 'w') as f:
        for item in strings_to_write:
            f.write("%s\n" % item)
```
